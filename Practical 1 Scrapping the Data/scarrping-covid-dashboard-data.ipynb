### Install & Import Dependencies

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

### Setup Chrome Driver

```python
driver = webdriver.Chrome("chromedriver")
```

```python
states = []
total_covid_cases = []
total_recoverd_cases = []
vaccinated_people = []
```

### Set Website 

```python
driver.get("https://www.covid19india.org/")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
```

### Fetching specific table data

```python
table = soup.find('div',attrs={'class':'table-container'})
```

```python
data = []
row_data = []
for index , item in enumerate(table.find_all('div' , attrs={'class' : 'total'})):
    if index%6 == 0 and index != 0:
        data.append(row_data)
        row_data = []
    row_data.append(item.text)
```

```python
for item in table.find_all('div',attrs={'class':'state-name'}):
    states.append(item.text)
```

```python
for d in data:
    total_covid_cases.append(d[0])
    total_recoverd_cases.append(d[2])
    vaccinated_people.append(d[-1])
```

```python
states = states[:-1] # there is one extra entry so we remove it.
```

### Make a DataFrame Object

```python
data = pd.DataFrame({'States':states , 'Total Covid Cases':total_covid_cases,'Total Recovered Cases':total_recoverd_cases,"Vaccine Doses Administered":vaccinated_people})
```

### Convert Dataframe Object to CSV

```python
data.to_csv('covid-report-statewise.csv')
```
